```python
import numpy as np # biblioteca para fazer cálculos matemáticos
from sklearn.linear_model import LinearRegression # importa da biblioteca sklearn o modelo de Regressão Linear
import matplotlib.pyplot as plt # biblioteca para plotar os gráficos
import pandas as pd
```

```python
# EXERCÍCIO 1 Criar um script para prever o nível de um tanque com vazão de 10,20,30,5,35,40 m3/ min
# criando o conjunto de dados
vazao = np.array([10,20,30,5,35,40]).reshape(-1,1) # cria o conjunto de dados
nivel = np.array([50,60,70,45, 80, 85]) # conjunto de dados para o nível
# cria modelo e treina o modelo de regressão linear
modelo = LinearRegression() # armazena a regressão linear da variável modelo
modelo.fit(vazao,nivel) # treina o modelo de regressão linear
```

```python
# Teste de modelo criado
vazao_teste = np.array([10,20,30,5,35,40]).reshape(-1,1) # vazão de água em m3/min
# Fazer previsões com base na vazão água
nivel_previsto = modelo.predict(vazao_teste)
```

```python
# Imprimir as previsões
for i in range(len(vazao_teste)): # laço de repetição para exibir os dados
    # de saida
    print(f"Vazão: {vazao_teste[i][0]} m3/min - Nível previsto: {nivel_previsto[i]} metros")
# gráfico de pontos - scatter
plt.scatter(vazao, nivel, color='blue', label='Dados de Treinamento') 
# gráfico de linhas 
plt.plot(vazao_teste,nivel_previsto, color  = 'red',label='Dados de teste')
plt.xlabel('Vazão de Água (m3/min)')
plt.ylabel('Nível do Tanque (metros)')
plt.title('Relação entre Vazão de Água e Nível do Tanque')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# EXERCÍCIO 2 Criar um script para prever a temperatura de um forno a partir da quantidade de 
#calor fornecida.
#Q = [100,200,300,400,500,600,700]
#T = [30,35,40,45,60,100,150]

# Criando o conjunto de dados
calor = np.array([100, 200, 300, 400, 500, 600, 700]).reshape(-1, 1)  # Quantidade de calor fornecida
temperatura = np.array([30, 35, 40, 45, 60, 100, 150])  # Temperatura do forno correspondente

# Criando e treinando o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(calor, temperatura)
```

```python
# Teste do modelo criado
calor_teste = np.array([25, 50, 75, 100, 125, 150]).reshape(-1, 1)  # Quantidade de calor em teste
temperatura_prevista = modelo.predict(calor_teste)
```

```python
# Imprimir as previsões
for i in range(len(calor_teste)):
    print(f"Quantidade de Calor: {calor_teste[i][0]} unidades - Temperatura prevista: {temperatura_prevista[i]:.2f} °C")

# Gráfico de pontos (scatter)
plt.scatter(calor, temperatura, color='blue', label='Dados de Treinamento')

# Gráfico de linhas (regressão)
plt.plot(calor_teste, temperatura_prevista, color='red', label='Linha de Regressão')
plt.xlabel('Quantidade de Calor')
plt.ylabel('Temperatura do Forno (°C)')
plt.title('Relação entre Quantidade de Calor e Temperatura do Forno')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# EXERCÍCIO 3 Prever o valor de um imovel com base na área
# A=[50,100,120,300]
# Valor = [180.000,300.000,375.000,600.000]

# Criando o conjunto de dados
area = np.array([50, 100, 120, 300]).reshape(-1, 1)  # Área dos imóveis em m²
valor = np.array([180000, 300000, 375000, 600000])  # Valor dos imóveis em reais

# Criando e treinando o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(area, valor)
```

```python
# Teste do modelo criado
area_teste = np.array([80, 150, 250]).reshape(-1, 1)  # Áreas dos imóveis para teste
valor_previsto = modelo.predict(area_teste)
```

```python
# Imprimir as previsões
for i in range(len(area_teste)):
    print(f"Área: {area_teste[i][0]} m² - Valor previsto: R${valor_previsto[i]:,.2f}")

# Gráfico de pontos (scatter)
plt.scatter(area, valor, color='blue', label='Dados de Treinamento')

# Gráfico de linhas (regressão)
plt.plot(area_teste, valor_previsto, color='red', label='Linha de Regressão')
plt.xlabel('Área do Imóvel (m²)')
plt.ylabel('Valor do Imóvel (R$)')
plt.title('Relação entre Área do Imóvel e Valor')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# EXERCÍCIO 4 Prever os gastos dos clientes com base no salário.
# Renda = [1200,5000,8000,10000,30000]
# Gastos = [750,1500,4000,6000,18000]

# Criando o conjunto de dados
renda = np.array([1200, 5000, 8000, 10000, 30000]).reshape(-1, 1)  # Renda dos clientes
gastos = np.array([750, 1500, 4000, 6000, 18000])  # Gastos dos clientes

# Criando e treinando o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(renda, gastos)
```

```python
# Conjunto de dados de teste
renda_teste = np.array([2000, 6000, 12000, 25000]).reshape(-1, 1)  # Novos salários para previsão
gastos_previsto = modelo.predict(renda_teste)
```

```python
# Imprimir as previsões
for i in range(len(renda_teste)):
    print(f"Renda: R${renda_teste[i][0]:,.2f} - Gastos previstos: R${gastos_previsto[i]:,.2f}")

# Gráfico de pontos (scatter)
plt.scatter(renda, gastos, color='blue', label='Dados de Treinamento')

# Gráfico de linhas (regressão)
plt.plot(renda, modelo.predict(renda), color='red', label='Linha de Regressão')

plt.xlabel('Renda (R$)')
plt.ylabel('Gastos (R$)')
plt.title('Relação entre Renda e Gastos')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# EXERCÍCIO 5 Prever os gastos dos clientes com base no salário.
# Renda = [1200,5000,8000,10000,30000]
# Gastos = [750,1500,4000,6000,18000]

# Criando o conjunto de dados
consumo = np.array([50, 140, 30, 75, 123]).reshape(-1, 1)  # Consumo de energia em KW
demanda = np.array([100, 200, 80, 100, 160])  # Demanda de energia correspondente

# Criando e treinando o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(consumo, demanda)
```

```python
# Conjunto de dados de teste
consumo_teste = np.array([60, 120, 90, 150]).reshape(-1, 1)  # Novos consumos para previsão
demanda_previsto = modelo.predict(consumo_teste)
```

```python
# Imprimir as previsões
for i in range(len(consumo_teste)):
    print(f"Consumo: {consumo_teste[i][0]} KW - Demanda prevista: {demanda_previsto[i]:.2f} KW")

# Gráfico de pontos (scatter)
plt.scatter(consumo, demanda, color='blue', label='Dados de Treinamento')

# Gráfico de linhas (regressão)
plt.plot(consumo, modelo.predict(consumo), color='red', label='Linha de Regressão')

plt.xlabel('Consumo de Energia (KW)')
plt.ylabel('Demanda de Energia (KW)')
plt.title('Relação entre Consumo e Demanda de Energia')
plt.legend()
plt.grid(True)
plt.show()
```
