```python
import pandas as pd # importa a biblioteca pandas
import matplotlib.pyplot as plt # importa a matplotlib
```

```python
dados = pd.read_csv('imigrantes_canada.csv') # importa a base de dados
```

```python
dados.tail() # exibe as 4 ultimas linhas 
```

```python
dados.info() # exibe as informações do dataframe
```

```python
# seta o pais como index
dados.set_index(['País'],inplace=True) # seta a coluna País como index
anos = list(map(str,range(1980,2013))) # Transforma os anos em string utilizando função map
```

```python
anos # exibe como string 
```

```python
brasil = dados.loc['Brasil',anos] # filtro para pegar os dados somente do Brasil
```

```python
brasil
```

```python
# cria um dicionário com os dados do Brasil
brasil_dic={
    'ano':brasil.index.tolist(), # transforma a coluna com o index em uma lista
    'imigrantes': brasil.values.tolist() #transforma a coluna com os valores em uma lista
    }
```

```python
# criando datagrame para o Brasil
df_brasil = pd.DataFrame(brasil_dic)
```

```python
df_brasil.head() # exibe os valores
```

```python
# plotando o número de imigrantes do Brasil para o Canadá
plt.plot(df_brasil['ano'],df_brasil['imigrantes']) #plota o gráfico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14) # ajuste de valores para o eixo x
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores para o eixo y
plt.xlabel("Anos",fontsize=14) # titulo eixo x
plt.title('Imigração do Brasil para o Canadá',fontsize =18) # adiciona título no gráfico
plt.ylabel("Número de imigrantes",fontsize=14) # titulo eixo y
plt.show() #exibe a figura
```

```python
argentina = dados.loc['Argentina',anos] # filtro para pegar os dados somente da Argentina
```

```python
# cria um dicionário com os dados da Argentina
argentina_dic={
    'ano':argentina.index.tolist(), # transforma a coluna com o index em uma lista
    'imigrantes': argentina.values.tolist() #transforma a coluna com os valores em uma lista
    }
```

```python
# criando datagrame para o Brasil
df_argentina = pd.DataFrame(argentina_dic)
```

```python
# plotando o número de imigrantes do Argentina para o Canadá
plt.plot(df_argentina['ano'],df_argentina['imigrantes']) #plota o gráfico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14) # ajuste de valores para o eixo x
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores para o eixo y
plt.xlabel("Anos",fontsize=14) # titulo eixo x
plt.title('Imigração da Argentina para o Canadá',fontsize =18) # adiciona título no gráfico
plt.ylabel("Número de imigrantes",fontsize=14) # titulo eixo y
plt.show() #exibe a figura
```

```python
plt.plot(df_argentina['ano'],df_argentina['imigrantes'],label = 'Argentina')
plt.plot(df_brasil['ano'],df_brasil['imigrantes'],label = 'Brasil')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel('Anos',fontsize=14)
plt.legend()
plt.ylabel('Número de imigrantes',fontsize=14)
plt.title('Imigração da Argentina para o Canadá')
plt.savefig("DadosImigracao.jpg")
plt.show()
```

```python
plt.bar(df_brasil['ano'],df_brasil['imigrantes']) # plt bar grafico de barras
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize = 14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.title("Imigração do Brasil para o Canadá")
plt.show()
```

```python
df_comparacao = dados.loc[['Brasil','Argentina'],anos] #filtro que pega duas colunas do dataset
```

```python
df_comparacao = df_comparacao.T # matriz transposta
```

```python
df_comparacao.head()
```

```python
df_comparacao['T']=df_comparacao['Brasil']+ df_comparacao['Argentina'] # soma duas colunas do dataframe
```

```python
plt.figure(figsize=(10,10)) # tamanho da figura
plt.plot(df_comparacao['Brasil'],label = 'Brasil',color ='green')
plt.plot(df_comparacao['Argentina'],label= 'Argentina',color='blue')
plt.plot(df_comparacao['T'],label= 'Total',color='red')
plt.xlabel('Anos',fontsize=14)
plt.ylabel('Número de imigrantes',fontsize=14)
plt.legend(loc='upper left')
plt.title("Imigração do Brasil e \n Argetina para o Canadá")
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14)
plt.show()
```

```python
fig, ax  = plt.subplots(figsize=(10,8))  # cria a variável ax para armazenar o subPlot
ax.plot(df_brasil['ano'],df_brasil['imigrantes'],lw=3,color = 'green',label='Brasil') # comando para plotar o grafico
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # altera o intervalo do eixo x
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
ax.set_title('Imigração do Brasil para o Canadá  1980 a 2013',fontsize=18)
ax.set_xlabel('Ano',fontsize=14)
ax.set_ylabel('Nº de imigrantes',fontsize=14)
ax.legend()
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5)) # cria um variável axs
# que corresponde a eixos diferentes no gráfico
# subplot (qtde de linhas, qtde de colunas, figsize - tamanho da figura)
axs[0].plot(df_brasil['ano'], df_brasil['imigrantes']) # plota o primeiro gráfico na posição 0
axs[0].set_title("Imigrantes do Brasil para o Canadá de \n 1980  -  2013", fontsize = 18)
axs[0].set_xlabel('Anos',fontsize=12)
axs[0].set_ylabel('Número de imigrantes',fontsize = 12)
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5)) # Cria o intervalo de 5 em 5 no eixo x
axs[0].grid() #Cria o grid do gráfico

#Gráfico 2
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Box Plot de imigração do Brasil \n para o Canadá', fontsize=18)
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Nº do imigrantes',fontsize=12)
axs[1].grid()
plt.show()
```
