```python
import pandas as pd #importa a biblioteca pandas
import numpy as np # importa a biblioteca numpy
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
```

```python
Dados_x = np.linspace(0,10,10) # gera dados em x de 10 em 10 valores
Dados_y = Dados_x * Dados_x # multiplica os dados de x e armazena em y
```

```python
plt.figure(figsize=(10,8), facecolor='gray') # cria uma figura
Ax = plt.axes() # cria os eixos para o gráfico
Ax.set_facecolor('white') # cor do fundo do gráfico
plt.plot(Dados_x, Dados_y,
         color='red',marker='o',
         linestyle = 'dashed',
         markeredgecolor='black'
         ) # plota o gráfico em x e y 
plt.xlabel("Eixo x",fontsize=16)
plt.ylabel('Eixo y',fontsize=16)
plt.title('Gráfico simples com Matplotlib',fontsize= 20)
```

```python
plt.rcParams.update({'font.size':15}) #função que permite ajustar
plt.figure(figsize= (10,10)) # define um tamanho para a figura
linhas = 2 # cria 2 linhas 
colunas = 2 # cria 2 linhas
plt.subplot(linhas,colunas,1) # subplot para plotar graficos um ao lado do outro
plt.plot(Dados_x, 'r--', label='Dados_x') # label rótulo para os dados
plt.legend(loc = 'lower right') # cria legenda pro gráfico no canto inferior direito
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title('Grafico 1')
plt.subplot(linhas,colunas,2)
plt.plot(Dados_x, Dados_y, 'b-,', label='Dados_x \n e y') # plota o grafico alterando a cor da linha
plt.legend(loc = 'lower right') # cria legenda pro gráfico no canto inferior direito
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title('Grafico 2')
plt.subplot(linhas,colunas,3)
plt.plot(Dados_x * Dados_x, 'g-', label='Dados_x^2')
plt.legend(loc = 'lower right') # cria legenda pro gráfico no canto inferior direito
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title('Grafico 3')
plt.subplot(linhas,colunas,4)
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title('Grafico 4')
plt.plot(Dados_y, 'c--', label='Dados_y')
plt.legend(loc = 'lower right') # cria legenda pro gráfico no canto inferior direito
plt.show() # remove o warning
```

```python
# gráfico de barras
Figura = plt.figure() # Cria uma variável chamada figura
Grafico = Figura.add_axes([0,0,0.7,0.7]) # cria os eixos para o gráfico
# Cria rotulos
Rotulo = ['A','B','C','D','E','F','G','H','I','J']
# Plota o gráfico
Grafico.bar(Rotulo,Dados_x,color ='red')
```

```python
Grafico.bar(Rotulo,Dados_x,color='red') # dados 1
Grafico.bar(Rotulo,Dados_y*0.1,color='blue')# dados 2
Figura
```
